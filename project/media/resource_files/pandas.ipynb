```python
#get me the sum of the sale values of each product name sold between Feb 15, 2024 and Feb 21, 2024
#my abomination


price = df['product_price']
quantity = df['line_item_quantity']
sales = price * quantity


feb = pd.to_datetime(df['transaction_created_at']).dt.month
date = pd.to_datetime(df['transaction_created_at']).dt.day

#df.groupby(by=['product_name']).sum()yes

def screen(feb, date):
    if feb == 2 and date >= 15 and date <= 21:
        return df.groupby(by=['product_name']).sum()[sales]

df['product_name'].apply(screen)
```
